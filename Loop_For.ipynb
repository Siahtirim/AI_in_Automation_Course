```python
# loop in loop
for i in range(0, 4):
    print(f"First Loop {i}")
    for j in range(0, 6):
        print(f"Sec Loop{j}")
        for k in range(0, 6):
            print(f"Third Loop{k}")
```

```python
#break
for i in range(0 , 5):

    if i == 3:
        break
    print(f"First Loop {i}")
print("Break is done")
```

```python
#continue
for i in range(0, 5):
    if i == 3:
        continue
    print(i)
```

```python
#pass
for i in range(0, 5):
    if i == 3:
        pass
    print(i)
```

```python
num = 0
while True:
    print(f"I'm still alive {num}")
    num += 1
print("while is done")
```
