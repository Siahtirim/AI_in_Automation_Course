```python
#if
color = input("What is your favorite color? ")
if color == "red":
    print("Your color is red")
elif color == "blue":
    print("Your color is blue")
elif color == "green":
    print("Your color is green")
else:
    print("we can't recognize this color")
```

```python
# if for list
school = ["Ali", "Bob", "Charlie", "David"]
name = input("What is your name? ")
if name in school:
    print(f"{name} is there")
else:
    print(name,"is not there")
```

```python
#Logic in iF
x1 = True
x2 = False
school = ["Ali", "Bob", "Charlie", "David"]
name = input("What is your name? ")
if name in school and x1:
    print("It's true")
else:
    print("It's false")
```
